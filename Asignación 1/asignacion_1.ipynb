```python
def ciclo(inicio, final, Dia):
    for i in range(inicio, final + 1):
        if i == Dia:
            return True
    return False

def evaluar_Dia(Dia):
    inicio = 0
    final = 8
    for i in range(15):
        if ciclo(inicio, final, Dia):
            polinomio = i
            return inicio, final, polinomio
        else:
            inicio = final - 2
            final = inicio + 8
            if final > 87:
                final = 86
                inicio = 86 - 9
    #print(inicio, final, polinomio)
    return "El día a evaluar no se encuentra dentro de los datos"
evaluar_Dia(8)
```

```python
#función para optener el polinomía de sensor en el día específici pedido
def pol_Sensor(Dia, columna):
    intervalo = evaluar_Dia(Dia)
    px = []
    x = 0
    P = lagrange(np.arange(0, intervalo[1]- intervalo[0]), Comparar[columna][intervalo[0]:intervalo[1]])
    for i in range(0, (intervalo[1]-intervalo[0]-1)*10):
        px.append(P(x))
        x = x + 0.1
    print(P, intervalo)
    return P, px, intervalo
#pol_Sensor(5)
```

```python
# Algoritmo para la calibración
# une todos los valores repetidos para AMB y el Sensor IOT 
def unificar():
    X = []
    Y = []
    for i in range(len(Comparar['Sensor IOT'])):
        if Comparar['Sensor IOT'][i] in X:
            continue
        else:
            X.append(Comparar['Sensor IOT'][i])
        if Comparar['AMB'][i] in Y:
            continue
        else:
            Y.append(Comparar['AMB'][i])
    return X, Y

def diferencias():
    D = []
    for i in range(len(X)):
        D.append(Y[i] - X[i])
    print(len(D))
    fix, ax = mpl.subplots(figsize=(15,6))
    mpl.scatter(Comparar.index, D, color = 'red', alpha = 0.5)
    pl.xlabel('Datos AMB')
    pl.ylabel('Desviacion entre Sensor IOT y AMB')
    pl.plot(Comparar.index, D)
    pl.show()
    Comparar['Sensor IOT'][7:15]
    return D

numeros = [16, 4, 9, 1, 3, 20, 8]
def ordenar(F, D, Y):
    X_1 = F
    orden = F
    orden.sort()
    D_1 = D
    Y_1 = Y
    bandera = True
    i = 0
    while bandera:
        if F[i] > F[i+1]:
            X_1[i] = F[i+1]
            X_1[i+1] = F[i]
            D_1[i] = D[i+1]
            D_1[i+1] = D[i]
            Y_1[i] = Y[i+1]
            Y_1[i+1] = Y[i]
        if F[len(F)-i-1] < F[len(F)-i-2]:
            X_1[len(F)-i-1] = F[len(F)-i]
            X_1[len(F)-i] = F[len(F)-i-1]
            D_1[len(F)-i-1] = D[len(F)-i]
            D_1[len(F)-i] = D[len(F)-i-1]
            Y_1[i] = Y[i+1]
            Y_1[i+1] = Y[i]
        i = i + 1
        if i >= (len(F)-2):
            i = 0
        if orden == X_1:
            bandera = False
            print('yo los ordené')
            return X_1, D, Y
        F = X_1
        D = D_1
        Y = Y_1

def promedio(X,num):
    inicio = 0
    final = int(len(X)/num)
    prom = []
    for i in range(0, num+1):
        suma = 0
        for j in range(inicio, final):
            if j > (len(X)-1):
                prom.append(float(suma/int(j - inicio)))
                return prom
            suma = suma + X[j]
            #print(j)
        prom.append(float(suma/int(len(X)/num)))
        inicio = final
        final = inicio + int(len(X)/num)
    return prom

[X, Y] = unificar()
D = diferencias()
[X, D, Y] = ordenar(X, D, Y)
prom_D = promedio(D,7)
prom_X = promedio(X,7)
(prom_X[5]-prom_X[0])/50
prom_X[5]
```
